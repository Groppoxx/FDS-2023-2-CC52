**VISUALIZAR LOS DATOS**

* Histograma de las vistas de los videos

```python
# Histograma de las vistas de los videos
plt.figure(figsize=(12, 6))
sns.histplot(dfMXvideos['views'], bins=50, kde=True, color='skyblue')  # Color opcional para personalizar la visualización
plt.title('Distribución de Vistas de Videos en YouTube')
plt.xlabel('Número de Vistas')
plt.ylabel('Frecuencia')
plt.legend(['Distribución de Vistas'])

# Tabla de datos
tabla_vistas = dfMXvideos['views'].describe().reset_index()
tabla_vistas.columns = ['Estadística', 'Valor']
tabla_vistas = tabla_vistas.round(2)

ax = plt.gca()
ax.table(cellText=tabla_vistas.values,
         colLabels=tabla_vistas.columns,
         cellLoc='center',
         loc='bottom', bbox=[0, -0.7, 1, 0.5])

plt.show()
```

* Dispersión de likes y dislikes por categoría

```python
# Gráfico de dispersión de likes y dislikes por categoría
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='likes', y='dislikes', data=dfMXvideos, hue='category_id', palette='viridis', alpha=0.7)
plt.title('Dispersión entre Likes y Dislikes por Categoría')
plt.xlabel('Número de "Me gusta"')
plt.ylabel('Número de "No me gusta"')
scatter_plot.legend(title='Categoría', bbox_to_anchor=(1, 1))

# Tabla de datos
tabla_likes_dislikes = dfMXvideos.groupby('category_id')[['likes', 'dislikes']].mean().reset_index()
tabla_likes_dislikes.columns = ['Categoría', 'Promedio Likes', 'Promedio Dislikes']
tabla_likes_dislikes = tabla_likes_dislikes.round(2)

ax = plt.gca()
ax.table(cellText=tabla_likes_dislikes.values,
         colLabels=tabla_likes_dislikes.columns,
         cellLoc='center',
         loc='bottom', bbox=[0, -1.2, 1, 1])

plt.show()
```

* Histograma de la distribución de categorías

```python
# Histograma de la distribución de categorías
plt.figure(figsize=(12, 6))
hist_plot = sns.histplot(dfMXvideos['category_id'], bins=30, kde=True, color='skyblue')
plt.title('Distribución de Categorías')
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
hist_plot.legend(['Distribución de Categorías'])

# Tabla de datos
tabla_categorias = dfMXvideos['category_id'].value_counts().reset_index()
tabla_categorias.columns = ['Categoría', 'Frecuencia']

ax = plt.gca()
ax.table(cellText=tabla_categorias.values,
         colLabels=tabla_categorias.columns,
         cellLoc='center',
         loc='bottom', bbox=[0, -1.2, 1, 1])

plt.show()
```

* Nube de Palabras para Etiquetas (Tags)

```python
all_tags = ' '.join(dfMXvideos['tags'].dropna())

# Nube de palabras para Etiquetas
plt.figure(figsize=(12, 6))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_tags)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras para Etiquetas (Tags)')

plt.show()
```

* Gráfico de Barras de Canales más Populares


```python
top_channels = dfMXvideos['channel_title'].value_counts().nlargest(10)

# Gráfico de Barras de Canales más Populares
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x=top_channels.values, y=top_channels.index, palette='viridis')
plt.title('Top 10 Canales con Mayor Número de Videos en Tendencia')
plt.xlabel('Número de Videos en Tendencia')
plt.ylabel('Canales')
bar_plot.legend(['Número de Videos'])

# Tabla de datos
tabla_canales = dfMXvideos['channel_title'].value_counts().reset_index().head(10)
tabla_canales.columns = ['Canal', 'Número de Videos']

ax = plt.gca()
ax.table(cellText=tabla_canales.values,
         colLabels=tabla_canales.columns,
         cellLoc='center',
         loc='bottom', bbox=[0, -1.2, 1, 1])

plt.show()
```

**Valores Nulos**

* Cantidad de valores nulos en cada columna

```python
total = dfMXvideos.isnull().sum().sort_values(ascending = False)
porcentaje_nulos = (dfMXvideos.isnull().sum() / dfMXvideos.isnull().count()).sort_values(ascending = False)
datos_faltantes = pd.concat([total, porcentaje_nulos], axis = 1, keys = ['Total', 'Porcentaje'])

print('-----------------------------------------------',
      'Los datos faltantes del Dataset son: ', total,
      '-----------------------------------------------',
      'El % de datos faltantes del dataset:',
      datos_faltantes[datos_faltantes['Total'] > 0], sep = '\n')
```

**Valores Atipicos**

---

* Cantidad de valores atipicos en cada columna

```python
numeric_columns = dfMXvideos.select_dtypes(include=['float64']).columns

outliers_info_list = []

for column in numeric_columns:
    Q1 = dfMXvideos[column].quantile(0.25)
    Q3 = dfMXvideos[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers_count = len(dfMXvideos[(dfMXvideos[column] < lower_limit) | (dfMXvideos[column] > upper_limit)])

    outliers_info_list.append({'Columna': column, 'Cantidad de Valores Atípicos': outliers_count})


pd.DataFrame(outliers_info_list)
```
